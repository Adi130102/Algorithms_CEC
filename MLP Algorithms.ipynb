```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression,LogisticRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder 
from sklearn.metrics import mean_squared_error,accuracy_score,precision_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
```

```python
csv=pd.read_csv("Salary_Data.csv")
# csv=csv.loc[:30]
csv
```

```python
newobject=LabelEncoder()
csv["Gender"]=newobject.fit_transform(csv["Gender"])
csv["Education Level"]=newobject.fit_transform(csv["Education Level"])
csv["Job Title"]=newobject.fit_transform(csv["Job Title"])
csv.dropna(inplace=True)
csv
```

```python
x=csv.drop(['Salary'],axis=1)
y=csv['Salary']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

# Linear Regression

```python
reg=LinearRegression()
reg.fit(x_train,y_train)
```

```python
pred_y=reg.predict(x_test)
pred_y
```

```python
mean_squared_error(y_test,pred_y)
```

# Logistic Regression

```python
log=LogisticRegression()
log.fit(x_train,y_train)
```

```python
pred_y=log.predict(x_test)
pred_y
```

```python
mean_squared_error(y_test,pred_y)
```

```python
accuracy_score(y_test,pred_y)
```

```python
precision_score(y_test,pred_y,average=None)
```

# KNN - K Nearest Neighbors

```python
knn=KNeighborsClassifier(n_neighbors=2)
knn.fit(x_train,y_train)
```

```python
pred_y=knn.predict(x_test)
pred_y
```

```python
mean_squared_error(y_test,pred_y)
```

```python
accuracy_score(y_test,pred_y)
```

```python
precision_score(y_test,pred_y,average=None)
```

# SVM - Simple Vector Machine

```python
svm=SVC()
```

```python
svm.fit(x_train,y_train)
```

```python
pred_y=svm.predict(x_test)
pred_y
```

```python
mean_squared_error(y_test,pred_y)
```

```python
accuracy_score(y_test,pred_y)
```

```python
precision_score(y_test,pred_y,average=None)
```

# Decision Tree

```python
tree=DecisionTreeClassifier()
tree.fit(x_train,y_train)
```

```python
pred_y=tree.predict(x_test)
pred_y
```

```python
mean_squared_error(y_test,pred_y)
```

```python
accuracy_score(y_test,pred_y)
```
