```python
import pandas as pd
import numpy as nm
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn import svm
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,precision_score,mean_squared_error
```

```python
data=pd.read_csv("Taxi_Platform.csv")
```

```python
data.head()
```

```python
data.shape
```

```python
data.dropna(inplace=True)
```

```python
data.shape
```

```python
data.info()
```

```python
data['satisfaction'].value_counts()
```

```python
data['Class'] = data['Class'].map({"Business": 3, "Eco Plus": 2, "Eco": 1}) 
```

```python
data['satisfaction'] = data['satisfaction'].map({"satisfied": 1, "dissatisfied": 0})
```

```python
data['Gender'].unique()
```

```python
data['Gender'] = data['Gender'].map({"Female": 1, "Male": 0})
```

```python
data['Customer Type'].value_counts()
```

```python
data['Customer Type']=data['Customer Type'].map({"Loyal Customer": 1, "disloyal Customer": 0})
```

```python
data['Type of Travel'].value_counts()
```

```python
data['Type of Travel']=data['Type of Travel'].map({"Business travel": 1, "Personal Travel": 0})
```

```python
data.info()
```

```python
print(data.keys())
```

```python
x=data.drop('Road Distance',axis=1)
y=data['Road Distance']
```

```python
x.head()
```

# linear regression

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.4)
```

```python
lm = LinearRegression()
lm.fit(X_train,y_train)
```

```python
y_pred=lm.predict(X_test)
```

```python
mean_squared_error(y_test,y_pred) 
```

# kNN

```python
x=data.drop('satisfaction',axis=1)
y=data['satisfaction']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4)
```

```python
knn=KNeighborsClassifier(n_neighbors=2)
knn.fit(x_train,y_train)
```

```python
y_pred=knn.predict(x_test)
```

```python
y_pred
```

```python
accuracy_score(y_test,y_pred)
```

# logistic regression


```python
log=LogisticRegression()
log.fit(x_train,y_train)
```

```python
y_pred=log.predict(x_test)
```

```python
y_pred
```

```python
accuracy_score(y_test,y_pred)
```

# SVM

```python
svm_model=svm.SVC(kernel='linear')
svm_model.fit(x_train[:100], y_train[:100])
```

```python
y_pred=svm_model.predict(x_test[:20])
```

```python
accuracy_score(y_test[:20],y_pred)
```

# decision tree


```python
dtree = DecisionTreeClassifier()
dtree.fit(x_train,y_train)
```

```python
y_pred=dtree.predict(x_test)
```

```python
y_pred
```

```python
accuracy_score(y_test,y_pred)
```
